```python
import nltk
import pandas as pd


df = pd.read_csv('stockerbot-export1.csv')

df.head()
```

**CLEANING THE TWEETS BY PURGING URLS, HYPERLINKS, SPECIAL CHARACTERS AND EXTRA WHITESPACES.**

```python
pat1 = r'@[A-Za-z0-9]+' # this is to remove any text with @....
pat2 = r'https?://[A-Za-z0-9./]+'  # this is to remove the urls
combined_pat = r'|'.join((pat1, pat2)) 
pat3 = r'[^a-zA-Z]' # to remove every other character except a-z & A-Z
combined_pat2 = r'|'.join((combined_pat,pat3))
```

```python
import re
import nltk

pat1 = r'@[A-Za-z0-9]+' # this is to remove any text with @....
pat2 = r'https?://[A-Za-z0-9./]+'  # this is to remove the urls
combined_pat = r'|'.join((pat1, pat2)) 
pat3 = r'[^a-zA-Z]' # to remove every other character except a-z & A-Z
combined_pat2 = r'|'.join((combined_pat,pat3))

dtweets = []

for i in range(0, len(df['text'])) : 
    w = df['text'][i]
    w.strip()
    tweets = re.sub(combined_pat2,' ',w)
    wtweets = re.sub(' +',' ',tweets)
    dtweets.append(wtweets)
df['Tweets'] = dtweets   
df['Tweets'].head()
```

**STEP 1 CASE FOLDING**

We lower the case of all the tweets using .lower().

```python
import string
string.punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

df['Tweets']= df['Tweets'].apply(lambda x:remove_punctuation(x))
df['Tweets']= df['Tweets'].apply(lambda x: x.lower())
df['Tweets'].head()
```

**STEP 2 TOKENIZE**

We tokenize the tweets and identify the tokens by using the regex library.

```python
import re
def tokenization(text):
    tokens = re.split('\W',text)
    return tokens
#applying function to the column
df['Tweet_token']= df['Tweets'].apply(lambda x: tokenization(x))
df[df.columns[8:]].head()
```

**STEP 3 STOPWORDS REMOVAL**

We remove the stopwords by using the the stopwords library to filter out the tweets.

```python
import nltk
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
df['Tweet_stopw']= df['Tweet_token'].apply(lambda x:remove_stopwords(x))
df[df.columns[8:]].head()
```

**STEP 4 STEMMING**

Stemming is done on the tweets using PorterStemmer from the NLTK Library

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
def stemming(text):
    stweets= [ps.stem(word) for word in text]
    return stweets
df['Tweet_stem']=df['Tweet_stopw'].apply(lambda x: stemming(x))
df[df.columns[8:]].head()
```

**STEP  5 LEMMATIZATION**

Lemmatization is perform using the inbuilt library WordNetLemmatizer.

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
def lemm(text):
    ltweets = [lemmatizer.lemmatize(word) for word in text]
    return ltweets
df['Tweet_lem']=df['Tweet_stopw'].apply(lambda x:lemm(x))
df[df.columns[8:]].head()
```

**STEP 6  INVERTED INDEX**

Finally we create an inverted index that maps the words to its location in the document.

```python
dict = {}
  
for i in range(0,len(df['Tweet_stopw'])):
    check = df['Tweets'][i].lower()
    for item in df['Tweet_stopw'][i]:
  
        if item in check:
            if item not in dict:
                dict[item] = []
  
            if item in dict:
                dict[item].append(i+1)
dict['video']
dict['office']
```
